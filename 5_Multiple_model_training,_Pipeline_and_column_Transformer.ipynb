```python
## Since Time is a nominal variable , we will use label encoder
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['time'] = encoder.fit_transform(df['time'])
df
```

```python
X = df.drop('time',axis = 1)
y = df['time']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=1)
```

```python
## Handling the missing value
# data encoding
## Featuure scaling


from sklearn.impute import SimpleImputer ## for missing value
from sklearn.preprocessing import OneHotEncoder## for encoding the data
from sklearn.preprocessing import StandardScaler ## for scaling

from sklearn.pipeline import Pipeline  #A sequence of data transformers
from sklearn.compose import ColumnTransformer #Groups all the pipeline steps for each of the clumns
```

```python
cat_cols = ["sex","smoker","day"]
num_cols = ["total_bill","tip","size"]
```

```python
## Fteature engineering automation using pipeline adn columntransfer

num_pipeline  = Pipeline(steps=[('imputation',SimpleImputer(strategy='median')),

                                ('scaling',StandardScaler())])

cat_pipeline  = Pipeline(steps=[('imputation',SimpleImputer(strategy='most_frequent')),

                                ('encoding',OneHotEncoder())])
```

```python
preprocessor = ColumnTransformer([('num_pipeline',num_pipeline,num_cols),
                   ('cat_pipeline',cat_pipeline,cat_cols)])
```

```python
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
model = {"support vector classifier":SVC(),
         "DT classifier":DecisionTreeClassifier()}
```

```python
from sklearn.metrics import accuracy_score
def model_train_eval(X_train,y_train,X_test,y_test,model):
  evaluation ={}
  for i in range(len(model)):
    model = list(model.values())[i]
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    model_score =accuracy_score(y_test,y_pred)
    evaluation[list(model.keys())[i]] = model_score
  return evaluation

```

```python
model_train_eval(X_train,y_train,X_test,y_test,model)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

models = {
    "Support Vector Classifier": SVC(),
    "Decision Tree Classifier": DecisionTreeClassifier()
}

def model_train_eval(X_train, y_train, X_test, y_test, models):
    evaluation = {}
    for name, clf in models.items():
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        model_score = accuracy_score(y_test, y_pred)
        evaluation[name] = model_score
    return evaluation

# Example usage
model_train_eval(X_train, y_train, X_test, y_test, models)
```
