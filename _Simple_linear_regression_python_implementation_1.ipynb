Linear Regression models the relationship between the dependent variable (y) and one or more independent variables (x) by fitting a linear equation to observed data. In simple terms, it finds the best-fitting straight line through the data points

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Read the data
## here we are loading data from sklearn
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
diabetes
```

```python
print(diabetes.DESCR)
```

```python
data = pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
```

```python
data['target']= diabetes.target
```

```python
data
```

```python
### EDA,DATA CLEANING ,DATA PREPPARATION,FEATURE ENGINEERING
```

```python
## DIVIDE THE DATA INTO X AND Y ,for simple linear regression we only take eone feature

X = data[['bmi']]
y = data['target']
```

```python
X
```

```python
y
```

```python
## split the data into Train Test -split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state= 1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
data.shape
```

```python
y_train
```

```python
y_test
```

```python
## Scaling (optional)
```

```python
## model traning

from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
## Instanceciate an LR model
model
```

```python
## Traninig the model
model.fit(X_train,y_train)## At back hand it will learn use gradient descent to give optimal coeffecient
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_predict = model.predict(X_test)
```

```python
y_predict
```

```python
## visualizr the result
plt.scatter(X_test,y_test,color = 'black',label = "Actual data")
plt.plot(X_test,y_predict,color = 'blue',linewidth = 3 ,label ="linear Regression line")
plt.xlabel("BMI")
plt.ylabel("One year progression -target")
plt.title("Linear regression on diabetes data")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_predict)
```
