```python
## data interpolation >> process of estimating data between given range
```

```python
## liner interpolation
## cubiic interpolation
## polynomial interpolation
```

```python
## liner interpolation
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

```

```python
plt.scatter(x,y)
```

```python
x_new = np.linspace(1,5,10)
```

```python
x_new
```

```python
y_interp = np.interp(x_new,x,y)
```

```python
y_interp
```

```python
plt.scatter(x_new,y_interp)
```

```python
## cubic interpolation

x = np.array([1,2,3,4,5])
y =  np.array([1,8,27,64,125])

plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
```

```python
f  = interp1d(x,y,kind = 'cubic')
```

```python
x_new = np.linspace(1,5,10)
y_interp = f(x_new)
```

```python
y_interp
```

```python
plt.scatter(x_new,y_interp)
```

```python
##ploynomial interpolation

x = np.array([1,2,3,4,5])
y = np.array([1,2,4,1,25])
```

```python
## interpolate using polynomial interpolation 
p = np.polyfit(x,y,2)
```

```python
x_new = np.linspace(1,5,10)
y_interp = np.polyval(p,x_new)
```

```python
y_interp
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```
