```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset("mpg")
```

```python
df.head()
```

```python
df.drop("name",axis =1,inplace=True)
```

```python
df.isna().sum()
```

```python
(df['horsepower'].median())
```

```python
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.dtypes
```

```python
df['origin'].value_counts()
```

```python
df['origin'] = df['origin'].map({"usa":1,"japan" :2 ,"europe":3})
```

```python
df['origin'] = df['origin'].astype(int)
```

```python
df.dtypes
```

```python
##seprate x and y
X = df.drop('mpg',axis = 1)
y =df['mpg']
```

```python
X
```

```python
y
```

```python
## Train test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
X_train.shape,X_test.shape
```

```python
### Simple linear regression model
from sklearn.linear_model import LinearRegression

regression_model = LinearRegression()
```

```python
regression_model
```

```python
regression_model.fit(X_train,y_train)
```

```python
for i, col_name in  enumerate(X_train.columns):
  print(i,col_name)
```

```python
for i, col_name in  enumerate(X_train.columns):
  print(f"The coeffficient for {col_name} is {regression_model.coef_[i]}")

```

```python
## coefficiets are relatively smaller, if one independent varaible changes
## slighttly then there will be no much difference in prediction
## This is sometime called as smoother model
```

```python
from sklearn.metrics import r2_score

y_pred_linear = regression_model.predict(X_test)
r2_linear = r2_score(y_test,y_pred_linear)
print(f"R square of linear regression {r2_linear}")
```

```python
## Ridge regression
from sklearn.linear_model import Ridge
ridge_regression_model = Ridge(alpha=0.1)
ridge_regression_model.fit(X_train,y_train)

for i, col_name in  enumerate(X_train.columns):
  print(f"The coeffficient for {col_name} is {ridge_regression_model.coef_[i]}")

```

```python
y_pred_linear = regression_model.predict(X_test)
r2_ridge = r2_score(y_test,y_pred_linear)
print(f"R square of Ridge regression {r2_linear}")
```

```python
## We dont see much varaition in coeffciet of ridge regression as copmared to linear regression
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso_regres_model = Lasso(alpha=0.5)
lasso_regres_model.fit(X_train,y_train)

for i, col_name in  enumerate(X_train.columns):
  print(f"The coeffficient for {col_name} is {lasso_regres_model.coef_[i]}")


```

```python
## # features coeffcient are 0 , lasso helps in feature selection
```

```python
y_pred_lasso =lasso_regres_model.predict(X_test)
r2_lasso = r2_score(y_test, y_pred_lasso)
print(f"R-squared score for Lasso Regression: {r2_lasso}")
```

```python
from sklearn.linear_model import ElasticNet
elastic_net_model = ElasticNet(alpha=1,l1_ratio=0.5)
elastic_net_model.fit(X_train,y_train)
```

```python
for i, col_name in  enumerate(X_train.columns):
  print(f"The coeffficient for {col_name} is {elastic_net_model.coef_[i]}")
```

```python
# Predict on the test set
y_pred_elastic_net = elastic_net_model.predict(X_test)

# Calculate evaluation metrics
r2_elastic_net = r2_score(y_test, y_pred_elastic_net)


print(f"R-squared score for Elastic Net Regression: {r2_elastic_net}")
```

```python
from sklearn.linear_model import RidgeCV
ridgecv=RidgeCV(cv=5)
ridgecv.fit(X_train,y_train)
y_pred=ridgecv.predict(X_test)

score=r2_score(y_test,y_pred)
print("R2 Score", score)
```
