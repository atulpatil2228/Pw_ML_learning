```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generatic synthetic data points
np.random.seed(1)
```

```python
## generating synthetic X
X = 2*np.random.rand(100,1)
```

```python
X
```

```python
## Gnerating synthetic dp y
y = 4+3*X+1.5*X**2+np.random.randn(100,1)
```

```python
y
```

```python
## train test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
X_train,X_test
```

```python
## performm polynominal regression
from sklearn.preprocessing import PolynomialFeatures
```

```python
degree = 2
poly_features = PolynomialFeatures(degree=2)
X_poly_train= poly_features.fit_transform(X_train)
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ploy_reg = LinearRegression()
```

```python
ploy_reg
```

```python
ploy_reg.fit(X_poly_train,y_train)
```

```python
ploy_reg.coef_
```

```python
ploy_reg.intercept_
```

```python
## making prediction on train data itself
y_poly_predict = ploy_reg.predict(X_poly_train)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse_train = mean_squared_error(y_train, y_poly_predict)
print(f'Mean Squared Error on Training Data: {mse_train}')
```

```python
plt.scatter(X_train, y_train, label = "training data")
```

```python
plt.scatter(X_train,y_train,label = 'traning')
X_range = np.linspace(0, 2, 100).reshape(-1, 1)
X_range_poly = poly_features.transform(X_range)
plt.plot(X_range, ploy_reg.predict(X_range_poly), color='red', label=f'Polynomial Regression (Degree {degree})')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression')
plt.legend()
plt.show()
```
